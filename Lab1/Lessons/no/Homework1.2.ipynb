```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#a. A random matrix with size varying with n = {10, 20, 30, ..., 100}

n_vector = np.arange(10, 100, 10)
i = 0
k_2 = np.zeros(np.shape(n_vector))
k_inf = np.zeros(np.shape(n_vector))
error = np.zeros(np.shape(n_vector))

for n in n_vector:
    A = np.random.randn(n,n)
    x_true = np.ones((n,))

    b = A@x_true
    k_2[i] = np.linalg.cond(A, 2)
    k_inf[i] = np.linalg.cond(A, np.inf)
    x = np.linalg.solve(A, b)
    error[i] = np.linalg.norm(x-x_true, 2)/np.linalg.norm(x_true, 2)
    i=i+1

x = n_vector
y1 = error

y2 = k_2
y3 = k_inf

plt.plot(x, y1)
plt.title('Random matrix - error')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.plot(x, y2)
plt.plot(x, y3)
plt.title('Ranodm matrix - condition number')
plt.xlabel('x')
plt.ylabel('y')
plt.legend('k_2', 'k_inf')
plt.show()

    
```
